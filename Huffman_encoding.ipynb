# **Huffman Encoding** 
###### Used the skeleton code from Goodrich page 377 to create heap priority queue

```python
class PriorityQueue():
    
    def __init__ (self,initial_list):
        self.initial_list=initial_list
        if len(self.initial_list)==0:
            ""
        else:
            self.insert("")

    def parent_pos(self,pos):
        return((pos-1)//2)
    
    def swap_val(self,i,j):
        self.initial_list[i],self.initial_list[j]=self.initial_list[j],self.initial_list[i]

    def pos_left_lt_pos_right(self,left,right):        
        if self.initial_list[[left][0]]>self.initial_list[[right][0]]:
            return True
        else:
            return False
    
    def up_heap(self,pos):
        parent=self.parent_pos(pos)
        if pos>0 and self.pos_left_lt_pos_right(parent,pos):
            self.swap_val(parent,pos)
            self.up_heap(parent)
    
    def left_chld(self,pos):
        if (2*pos+1)<len(self.initial_list):
            return(2*pos+1)
        else:
            return("")
    
    def right_chld(self,pos):
        if (2*pos+2)<len(self.initial_list):
            return(2*pos+2)
        else:
            return("")


    def down_heap(self,pos):
        left_child=self.left_chld(pos)
        right_child=self.right_chld(pos)
        temp_small=''

        if left_child!='':
            val_left=left_child
            temp_small=val_left
            if right_child!='':
                val_right=right_child
                if self.pos_left_lt_pos_right(val_left,val_right):
                    temp_small=val_right
                else:
                    temp_small=val_left
        if temp_small=='':
            return
        if self.pos_left_lt_pos_right(pos,temp_small):
            self.swap_val(pos,temp_small)
            self.down_heap(temp_small)

    def insert(self,element):
        if type(element)==str:
            for i in range(len(self.initial_list)-1,-1,-1):
                self.down_heap(i)
        else:
            self.initial_list.append(element)
            self.up_heap(len(self.initial_list)-1)
        return

    def get_min (self):
        max_val=self.initial_list[0]
        last_pos_val=self.initial_list.pop()
        if len(self.initial_list)!=0:
            self.initial_list[0]=last_pos_val
            self.down_heap(0)
        return(max_val)
    
    def get_values(self):
        return(self.initial_list)

    def print (self):
        print(self.initial_list)
        return
```

##### Driver code for reading the file for which we need to carry out the coding. Creating a dictionary of the character frequency.

```python
def char_freq(filename):
    cntr_dict={}

    with open(filename,'r') as my_line:
        x=my_line.readlines()
        for i in x:
            for j in list(i):
                if cntr_dict.get(j,None) == None:
                    cntr_dict.update({j:1})
                else:
                    cntr_dict[j]+=1
    # print(cntr_dict)

    freq=cntr_dict.values()
    character=cntr_dict.keys()

    final_list=list(zip(freq,character))
    return(final_list)

```

##### Driver code that will help in creating the huffman codes

```python
def huff(my_str,code_table,flag):
    
    for j in list(my_str):
        # print(j)
        if code_table.get(j,None) == None:
            code_table.update({j:flag})
        else:
            temp=code_table.get(j)
            temp=temp+flag
            code_table.update({j:temp})
    return(code_table)
```

##### Driver file for creating huffman codes

```python
def create_codes(final_list):

    pq = PriorityQueue (final_list)
    tree_tuple=pq.get_values()
    code_table={}
    while len(tree_tuple)>1:
        min1=pq.get_min()
        code_table=huff(min1[1],code_table,str(0))
        min2=pq.get_min()
        code_table=huff(min2[1],code_table,str(1))
        new_freq=min1[0]+min2[0]
        new_str=min1[1]+min2[1]
        pq.insert((new_freq,new_str))
        tree_tuple=pq.get_values()

    final_code_table={}

    for i,j in code_table.items():
        final_code_table.update({i:j[::-1]})

    return(final_code_table)

```

##### **MASTER FUNCTION FOR IMPLEMENTING HUFFMAN CODING**

```python
def huff_coded_file(filename):
    
    final_list=char_freq(filename)
    final_code_table=create_codes(final_list)
    
    coded_str=''
    with open(filename,'r') as my_line:
        x=my_line.readlines()
        for i in x:
            for j in list(i):
                coded_str+=final_code_table.get(j)
    print("----------------The coded text which has been output to the output.txt file----------------")
    print(coded_str)
    with open('output.txt', 'w') as f:
        f.write(coded_str)
    return(final_code_table)


```
